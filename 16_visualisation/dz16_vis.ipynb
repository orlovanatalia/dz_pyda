#### Подготовка датасета для анализа.
Будут использованы данные об уровне счастья в разных странах за период 2016 - 2019 годы.
https://www.kaggle.com/unsdsn/world-happiness  
    Первичный анализ показывает, что данные для каждого года находятся в отдельном файле, названия и количество столбцов менялось от года к году. Поэтому необходимо определить параметры, которые являются общими для всех датасетов и создать единый датасет для построения визуализаций, который будет содержать значения каждого параметра для некольких лет. При объединении необходимо учитывать, что в разных таблицах один и тот же параметр может называться по-разному. 

```python
df2015 = pd.read_csv('2015.csv', sep=',')
df2016 = pd.read_csv('2016.csv', sep=',')
df2017 = pd.read_csv('2017.csv', sep=',')
df2018 = pd.read_csv('2018.csv', sep=',')
df2019 = pd.read_csv('2019.csv', sep=',')
```

В итоговом датасете будем использовать следующие столбцы (в скобках указаны различные названия из разных таблиц):  
- Overall rank (Happiness.Rank)
- Country or region (Country)
- Score (Happiness.Score, Happiness Score)
- GDP per capita (Economy..GDP.per.Capita., Economy (GDP per Capita))
- Healthy life expectancy (Health..Life.Expectancy., Health (Life Expectancy))
- Freedom to make life choices (Freedom)
- Generosity
- Perceptions of corruption (Trust..Government.Corruption., Trust (Government Corruption))  
Дополнительно в каждую таблицу добавим столбец "Year".

```python
df2019['Year'] = 2019
df2019.drop(['Social support'],axis=1, inplace=True)
df2019.head()
```

```python
df2018['Year'] = 2018
df2018.drop(['Social support'],axis=1, inplace=True)
df2018.head()
```

```python
df2015.info()
```

```python
df2015.rename(columns={'Country': 'Country or region', 
                       'Happiness Rank': 'Overall rank', 
                       'Happiness Score': 'Score',
                       'Economy (GDP per Capita)': 'GDP per capita',
                       'Health (Life Expectancy)': 'Healthy life expectancy',
                       'Freedom': 'Freedom to make life choices',
                       'Trust (Government Corruption)': 'Perceptions of corruption',
                      }, inplace=True)
df2015.drop(['Region', 'Family', 'Standard Error','Dystopia Residual'],axis=1, inplace=True)
df2015['Year'] = 2015
df2015.head()
```

```python
df2016.rename(columns={'Country': 'Country or region', 
                       'Happiness Rank': 'Overall rank', 
                       'Happiness Score': 'Score',
                       'Economy (GDP per Capita)': 'GDP per capita',
                       'Health (Life Expectancy)': 'Healthy life expectancy',
                       'Freedom': 'Freedom to make life choices',
                       'Trust (Government Corruption)': 'Perceptions of corruption',
                      }, inplace=True)
df2016.drop(['Region','Lower Confidence Interval', 'Upper Confidence Interval','Family','Dystopia Residual'],axis=1, inplace=True)
df2016['Year'] = 2016
df2016.head()
```

```python
df2017.rename(columns={'Country': 'Country or region', 
                       'Happiness.Rank': 'Overall rank', 
                       'Happiness.Score': 'Score',
                       'Economy..GDP.per.Capita.': 'GDP per capita',
                       'Health..Life.Expectancy.': 'Healthy life expectancy',
                       'Freedom': 'Freedom to make life choices',
                       'Trust..Government.Corruption.': 'Perceptions of corruption',
                      }, inplace=True)
df2017.drop(['Whisker.high','Whisker.low','Family','Dystopia.Residual'],axis=1, inplace=True)
df2017['Year'] = 2017
df2017.head()
```

```python
df = pd.concat([df2015, df2016, df2017, df2018, df2019])
df
```

```python
df.describe()
```

### График

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
dynamics_by_year = df[['Year', 'Score']].groupby('Year').mean()
dynamics_by_year
```

```python
dynamics_by_year['Score'].plot(title='Динамика уровня счастья', 
                                        xlabel='Год', 
                                        xticks = [2015,2016,2017,2018,2019],
                                        ylabel='Уровень')
```

Самый первый график - изменение среднего уровня счастья по всем странам за период наблюдений. Наблюдается снижение уровня в 2017 году и повышение к 2019 году. Скорее всего на такой усредненный параметр влияла общая ситуация в мире: в 2017 году была очень нестабильная ситуация на Ближнем Востоке, плюс эмигрантский кризис в Европе мог привести к небольшому спаду (-0,02). 2019 год с этой точки зрения был спокойнее. Скорее всего из-за пандемии Ковид-19 в 2020 году уровень опять снизится.Я

### Гистограмма

```python
df[df['Year'] == 2015].Score.plot(kind='hist', 
                     bins=18,
                     title='Score distribution',
                     xlabel='Score',
                     ylabel='Level',
                     alpha = 0.5,
                     density = True,
                     xticks=range(2, 8))
df[df['Year'] == 2019].Score.plot(kind='hist', 
                   bins=18,
                   title='Score distribution',
                   xlabel='Score',
                   ylabel='Level',
                   alpha = 0.5,
                   density = True,
                   xticks=range(2, 8))
```

Следующий график - гистограммы уровня счастья в 2015 и 2019 году. Видим, что  распределение немного отличается: в 2015 году его можно назвать волнообразным с пиками в районе 5, 6, 7, что говорит о неравномерности распределения этого параметра, а основе чего можно было бы выделить несколько групп стран по уровню счастья. Распределение 2019 года более сглаженное, но неравномерность все равно присутствует. Предполагаю, что это зависит от региона. Дальнейший анализ можно продолжить в направлении определения групп стран по уровню счастья.

### Boxplot

```python
import seaborn as sns
```

```python
sns.boxplot(x='Year', y='Healthy life expectancy',
            data=df)
```

Боксплоты по параметру "Ожидаемая продолжительность здоровой жизни" для разных лет показывают, что характер распределения (межквартильный интервал, длина "усов", положение медианы) практически не меняется от года к году. Характер изменений такой же как у общего уровня счастья: снижение в 2017 году и повышение в 2019, что является логичным, так человек будет чувствовать себя более счастливым, если будет считать, что проживет долгую жизнь.
